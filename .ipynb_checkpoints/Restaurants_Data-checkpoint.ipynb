```python
import pandas as pd
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func
from config import Password
import numpy as np
```

```python
restaurant_csv=pd.read_csv("./data/Datafiniti_Fast_Food_Restaurants.csv")
obesity_csv=pd.read_csv("./data/National-Obesity-Monitor-Data.csv")
restaurant_csv.head()
obesity_csv.head()
```

```python
obesity_df=obesity_csv
# obesity_df02=obesity_df.groupby("group")
obesity_df=obesity_df.drop(['1971-1974','1976-1980','1988-1994','1999-2000'],axis=1)
obesity_df.head()
```
